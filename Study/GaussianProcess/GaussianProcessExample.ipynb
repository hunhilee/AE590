# Gaussian Process

### Introduction

가우시안 프로세스를 이해하기 위해 regression 문제를 통해 알아보고자 한다. 일반적으로 regression 문제는 parametric model로 파라미터를 모델링하는 방법을 취하지만, 가우시안 프로세스는 non-parametric model으로 함수를 직접 모델링한다. 


### Problem

아래와 같은 식이 있다고 할때 이 식은 블랙박스로 되어 있고, 이 식에 대한 몇몇 observed data만 있다고 하자

$$f(x)=0.03 x^5 + 0.2 x^4 - 0.1 x^3 - 2.4 x^2 - 2.5 x + 6$$

그럼 몇몇 observed data만 가지고, 다음 식을 예측할 수 있을까.

```python
import numpy as np
import matplotlib.pylab as plt
```

우선 함수 $ f(x) $를 함수로 정의하고, 그래프로 표현한다. 이때 $X$ 범위가 너무 멀리 떨어지면 $Y$값이 너무 크므로, $X$의 범위는 $[-5, 3.5]$ 까지로 한다. 

```python
coefs = [6, -1.5, -2.4, -0.1, 0.2, 0.03]

def f(x):
    total = 0
    for exp, coef in enumerate(coefs):
        total += coef * (x ** exp)
    return total
```

```python
xs = np.linspace(-5, 3.5, 80)
ys = f(xs)
```

```python
plt.title("f(x)")
plt.plot(xs, ys, label='ground truth')
plt.legend()
plt.show()
```

다음으로 Training Data (Observed Data)를 설정한다. 여기에서는 4개의 데이터만 주어진다고 하자

```python
x_obs = np.array([-4, 0, 1.5, 2.7])
y_obs = f(x_obs)

print("Training Data Set")
for x,y in zip(x_obs, y_obs) : 
    print("X:{0:+0.2f}, Y:{1:+0.2f}".format(x, y))
```

다음으로 커널을 만들어 본다. 커널은 기본적으로 사용하는 **Squared Exponential**과 수업시간에 나온 커널을 만들어 비교해본다.

$$ \kappa1(x, x')=\exp\left(-~\frac{(x-x')^2}{2}\right)$$

$$ \kappa2(x, x')=\theta_0 \exp(-\frac{\theta_1}{2} \| x_n - x_m \|^2 ) + \theta_2 + \theta_3 x_n^T x_m $$

주의) $\theta_3$의 경우 $x_n$

```python
def k1(xs, ys, sigma=1, l=1):
    dx = np.expand_dims(xs, 1) - np.expand_dims(ys, 0)    
    return (sigma ** 2) * np.exp(-((dx / l) ** 2) / 2)

def m1(x):
    return np.zeros_like(x)

def k2(xn, xm, theta0 = 1., theta1 = 1., theta2 = 1., theta3 = 1.):
    dx = (np.expand_dims(xn, 1) - np.expand_dims(xm, 0))
#     return theta0 *  np.exp( - (theta1 / 2 ) * (dx**2)) + theta2 + theta3*xn.T*xm
    return theta0 *  np.exp( - (theta1 / 2. ) * (dx**2) ) + theta2 + theta3*np.outer(xn, xm)
 
def m2(x):
    return np.zeros_like(x)

```

우선 Training Data Set을 $ ( \mathbf{x},\mathbf{y} ) $이라고 정의하고, Test Data Set을 $ ( \mathbf{x}_*,\mathbf{y}_* ) $이라고 정의하자

우리가 원하는 $ \mathbf{y}_* $는 $ p(\mathbf{y}_* \vert \mathbf{x}_*, \mathbf{x}, \mathbf{y}) $으로 구할 수 있다.

이 conditioning multivariate gaussians을 구하기 위해서는 joint gaussians를 우선 알아야하고, $\mathbf{y}$과 $\mathbf{y}_*$의 joint gaussians을 아래와 같이 모델링 할 수 있다.

$$ \begin{pmatrix}\mathbf{y} \\ \mathbf{y}_*\end{pmatrix} \sim\mathcal{N}\left(\begin{pmatrix}m(\mathbf{x})\\ m(\mathbf{x}_*)\end{pmatrix}, \begin{pmatrix}\mathbf{K} & \mathbf{K}_* \\ \mathbf{K}_*^T & \mathbf{K}_{**}\end{pmatrix}\right) $$

where $\mathbf{K}=\kappa(\mathbf{x}, \mathbf{x})$, $\mathbf{K}_* = \kappa(\mathbf{x}, \mathbf{x}_*)$ and $\mathbf{K}_{**}=\kappa(\mathbf{x}_*, \mathbf{x}_*)$. 

joint gaussians를 활용하여, conditioning multivariate gaussians는 다음과 같이 구할 수 있다. 

$$\begin{align}
p(\mathbf{y}_*\vert \mathbf{x}_*, \mathbf{x}, \mathbf{y})&=\mathcal{N}(\mathbf{y}_*\vert \mu_*, \Sigma_*)\\
\mu_*&=m(\mathbf{x}_*)+\mathbf{K}_*^T\mathbf{K}^{-1}(\mathbf{y}-m(\mathbf{x}))\\
\Sigma_*&=\mathbf{K}_{**}-\mathbf{K}_*^T\mathbf{K}^{-1}\mathbf{K}_*
\end{align}$$


### GP using K1

첫번째로 수행해본 K1의 경우 놀랍게도 단지 4개의 학습 데이터 만으로도 regression이 잘 수행되었다.

```python
K = k1(x_obs, x_obs)
K_s = k1(x_obs, xs)
K_ss = k1(xs, xs)

K_sTKinv = np.matmul(K_s.T, np.linalg.pinv(K))

mu_s = m(xs) + np.matmul(K_sTKinv, y_obs - m(x_obs))
Sigma_s = K_ss - np.matmul(K_sTKinv, K_s)

stds = np.sqrt(Sigma_s.diagonal())
err_xs = np.concatenate((x_s, np.flip(x_s, 0)))
err_ys = np.concatenate((mu_s + 2 * stds, np.flip(mu_s - 2 * stds, 0)))

plt.title("GP with k1")
plt.plot(x_s, mu_s, label='estimated y_s', color='red')
plt.plot(xs, ys, label='ground truth', color='blue')
plt.fill(err_xs, err_ys, label='cov', color='green', alpha=0.3)
plt.legend()
plt.show()
```

### GP using K2

두번째로 수업시간에 배운 커널 함수를 가지고 수행을 해보았다. K1과 크게 다르지는 않아보이지만, 다양한 $theta$ 값을 변경할 수 있어 튜닝이 가능했다

```python
K = k2(x_obs, x_obs, theta0=1)
K_s = k2(x_obs, xs, theta0=1)
K_ss = k2(xs, xs, theta0=1)

K_sTKinv = np.matmul(K_s.T, np.linalg.pinv(K))

mu_s = m(xs) + np.matmul(K_sTKinv, y_obs - m(x_obs))
Sigma_s = K_ss - np.matmul(K_sTKinv, K_s)

stds = np.sqrt(Sigma_s.diagonal())
err_xs = np.concatenate((x_s, np.flip(x_s, 0)))
err_ys = np.concatenate((mu_s + 2 * stds, np.flip(mu_s - 2 * stds, 0)))

plt.title("GP with k2")
plt.plot(x_s, mu_s, label='estimated y_s', color='red')
plt.plot(xs, ys, label='ground truth', color='blue')
plt.fill(err_xs, err_ys, label='cov', color='green', alpha=0.3)
plt.legend()
plt.show()
```
